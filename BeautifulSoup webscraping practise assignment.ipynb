```python
soup=BeautifulSoup(page.content)
soup
```

```python
header_tags=[]
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://presidentofindia.nic.in/former-presidents')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
president_names=[]
```

```python
for i in soup.find_all('div',class_='desc-sec'):
    president_names.append(i.text)
```

```python
president_names
```

```python
import pandas as pd
df=pd.DataFrame({'Former_president_name':president_names})
df
```
