```python
# As there are outliers in most of the rows we will find  the Z-score, to identify the rows and columns with outliers.
from scipy.stats import zscore
dfzscore=np.abs(zscore(df))
print(np.where(dfzscore>3))
# rows and columns with Z-score more than +3 and less than -3 are considered as outliers
```

```python
dfnew=df[(dfzscore<3).all(axis=1)]
print('dfnew.shape=',dfnew.shape)
# this is the new dataframe after removing the rows and columns with Z score outside the permitted range
# the new dataframe consists of 1458 rows and 12 columns
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
# spliting the data into features and label. Independent variables will reagrded ad X( features) and dependent variable will be regarded as y(label)
```

```python
x
# Independent Variables
```

```python
y
# label/ dependent variable
```

```python
# Removing multicollinearity by checking the Variance Inflation Factor
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['VIF values']=[variance_inflation_factor(x.values,i)
                   for i in range(len(x.columns))]
vif['Features']=x.columns
vif
```

```python
#Columns with highest VIF and least correaltion with label will be removed
x.drop('density', axis=1, inplace=True)
x.drop('pH', axis=1, inplace=True)
# here we are removing density and pH coloum as they are showing very high VIF and have least cprrelation with Y
```

```python
x
# the density and pH columns have been dropped
# new dataframe consists of 1458 rows and 9 columns
```

```python
y.value_counts()
# target column is not balanced.
# so we need to use oversampling method to balance this
```

```python
# OVERSAMPLING
!pip install imblearn
from imblearn.over_sampling import SMOTE
SM=SMOTE()
x1,y1= SM.fit_resample(x,y)
```

```python
y1.value_counts()
# now the data is balanced and clean
```

```python
#MODELLING
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
maxAccu=0
maxRs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test= train_test_split(x1,y1,test_size=0.30, random_state=i)
    RFR= RandomForestClassifier()
    RFR.fit(x_train,y_train)
    pred= RFR.predict(x_test)
    acc= accuracy_score(y_test, pred)
    if acc>maxAccu:
        maxAccu= acc
        maxRs= i
print('Best Accuracy is:',maxAccu, 'at random_state',maxRs)
# The model is showing 95% accuracy at randon state 1
```

```python
# creating Train, Test, Split:
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.30,random_state=maxRs)
```

```python
# CLASSIFIACTION ALGORITHMS:
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
# RandomForest Classifier:
RFC= RandomForestClassifier()
RFC.fit(x_train,y_train)
predRFC=RFC.predict(x_test)
print(accuracy_score(y_test,predRFC))
print(confusion_matrix(y_test,predRFC))
print(classification_report(y_test,predRFC))
```

```python
#Logistic Regression
LR= LogisticRegression()
LR.fit(x_train,y_train)
predLR=LR.predict(x_test)
print(accuracy_score(y_test,predLR))
print(confusion_matrix(y_test,predLR))
print(classification_report(y_test,predLR))
```

```python
# Support Vector Machine Classifier:
svc=SVC()
svc.fit(x_train,y_train)
predsvc=svc.predict(x_test)
print(accuracy_score(y_test,predsvc))
print(confusion_matrix(y_test,predsvc))
print(classification_report(y_test,predsvc))
```

```python
# GradientBoosting Classifier:
GB=GradientBoostingClassifier()
GB.fit(x_train,y_train)
predGB=GB.predict(x_test)
print(accuracy_score(y_test,predGB))
print(confusion_matrix(y_test,predGB))
print(classification_report(y_test,predGB))
```

```python
#Adaboost Classifier:
ABC=AdaBoostClassifier()
ABC.fit(x_train,y_train)
predABC=ABC.predict(x_test)
print(accuracy_score(y_test,predABC))
print(confusion_matrix(y_test,predABC))
print(classification_report(y_test,predABC))
```

```python
#Bagging Classifier:
BG=BaggingClassifier()
BG.fit(x_train,y_train)
predBG=BG.predict(x_test)
print(accuracy_score(y_test,predBG))
print(confusion_matrix(y_test,predBG))
print(classification_report(y_test,predBG))
```

```python
#ExtraTreesClassifier:
ET=ExtraTreesClassifier()
ET.fit(x_train,y_train)
predET=ET.predict(x_test)
print(accuracy_score(y_test,predET))
print(confusion_matrix(y_test,predET))
print(classification_report(y_test,predET))
```
