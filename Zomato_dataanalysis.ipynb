```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_excel('zomato.xlsx')
data
```

```python
data.info()
# there are null values in 'Cuisines' column only
```

```python
data[data['Cuisines'].isnull()]
# to see the rows with null cuisine values
```

```python
data.Cuisines.value_counts()
```

```python
# filling the missing values in Cuisines column with NaN

data['Cuisines'] = data['Cuisines'].fillna('NaN')
```

```python
data.isnull().sum()
# no more missing values
```

```python
for col_name in data.columns:
    if(data[col_name].dtype == 'object'):
        data[col_name]= data[col_name].astype('category')
        data[col_name] = data[col_name].cat.codes
# changing all object data type to integer
```

```python
for col_name in data.columns:
    if(data[col_name].dtype == 'float'):
        data[col_name]= data[col_name].astype('category')
        data[col_name] = data[col_name].cat.codes
# changing all float datatypes to integer
```

```python
data.info()
# all the columns have same datatype now
```

```python
# reading the country-code data

df_country= pd.read_excel('Country-Code.xlsx')
df_country
```

```python
df_country.dtypes
```

```python
df_country.isnull().sum()
```

```python
# mergibg the two dataframes

df=pd.merge(data,df_country,on='Country Code',how='left')
df.head()
```

```python
ratings = df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
# India has the highest number of rating count
```

```python
plt.figure(figsize=(30,10))
sns.boxplot(x="Country", y="Rating color", data=df, dodge=True)
plt.show()
```

```python
country_online_delivery = df[["Has Online delivery", 'Country']].groupby(["Has Online delivery", 'Country']).size().reset_index(name='Count')
country_online_delivery
# online delivery facility is available only in India and USA
# India has the largest number of online delivering restaurants
```

```python
from scipy.stats import zscore
z_score=zscore(df[['Average Cost for two', 'Currency', 'Is delivering now', 'Has Table booking', 'Votes' ]])
abs_z_score= np.abs(z_score)                   
filter_entry= (abs_z_score <3).all(axis=1)
df= df[filter_entry]
df.reset_index(inplace=True)
```
